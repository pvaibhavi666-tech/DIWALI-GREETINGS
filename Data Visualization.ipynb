```python
# Bar plot or Bar graph
sns.barplot(x='species', y='sepal_length', data=df, hue='species')
plt.show()
```

```python
# Count plot
sns.countplot(x='species', data=df, hue='species')
plt.show()
```

```python
# Box plot
sns.boxplot(x='sepal_width', data=df)
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=df, hue='species')
plt.show()
```

```python
# Violin plot
sns.violinplot(x='species', y='sepal_length', data=df, hue='species')
plt.show()
```

```python
# Stripplot
sns.stripplot(x='species', y='sepal_length', data=df, hue='species')
plt.show()
```

```python
# Histogram
sns.histplot(x='sepal_length', data=df)
plt.show()
```

```python
sns.histplot(x='sepal_length', data=df, hue='species', kde=True)
plt.show()
# kde= lernel density estimation
```

```python
# Distplot
sns.displot(df['petal_width'], color="pink")
plt.show()
```

```python
# Pairplot
sns.pairplot(df, hue='species')
plt.show()
```

```python
# Categorical plot
sns.catplot(x='species', data=df, kind='violin')
plt.show()
```
